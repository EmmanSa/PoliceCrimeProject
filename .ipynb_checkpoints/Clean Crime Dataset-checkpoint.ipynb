```python
dataframes = []

for file in all_csv_files:
    try:
        df = pd.read_csv(file)
        
        # Clean column names. Add _ to names
        df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

        #Drop rows where 'crime_id' is missing
        df = df.dropna(subset=["crime_id"])

        # Drop rows where lat/long and location are missing
        df = df.dropna(subset=["latitude", "longitude", "location"], how="all")

        # Drop rows where LSOA name and code are missing
        df = df.dropna(subset=["lsoa_name", "lsoa_code"], how="all")

        # Add change format of date from YYYY-MM to YYYY-MM-01 as Snowflake does not accept YYYY-MM format
        df['month'] = pd.to_datetime(df['month'].astype(str) + "-01", errors='coerce')

        # Check list of police force for each area
        filename = os.path.basename(file).lower()
        for police_force, lsoa_prefixes in lsoa_names_list.items():
            if police_force in filename:
                #Filter to valid region
                df = df[df["lsoa_name"].str.contains('|'.join(lsoa_prefixes), case=False, na=False)]
                break

        # Drop any duplicate crime_ids
        df = df.drop_duplicates(subset=["crime_id"])

        if not df.empty:
            dataframes.append(df)

    except Exception as e:
        print(f" Error reading {file}: {e}")
```

```python
# Merge into one dataset
cleaned_df = pd.concat(dataframes, ignore_index=True)
```

```python
# Save final output
cleaned_df.to_csv("UK_Crime_Data.csv", index=False)
```
